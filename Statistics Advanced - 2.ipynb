

**Question 1: What is hypothesis testing in statistics?**

**Answer:**
Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. In simple terms, it helps us test whether an assumption (called a hypothesis) about a population is likely to be true or not.

There are two main types of hypotheses:

* **Null Hypothesis (H₀):** This is the default assumption that there is no effect, no difference, or no relationship. Example: “A new medicine has no effect on blood pressure.”
* **Alternative Hypothesis (H₁ or Ha):** This is what we want to prove, showing that there is some effect, difference, or relationship. Example: “The new medicine lowers blood pressure.”

**Steps in hypothesis testing:**

1. **State the hypotheses:** Define the null and alternative hypotheses.
2. **Choose a significance level (α):** Usually 0.05 (5%), meaning we allow a 5% chance of making an error.
3. **Collect and analyze data:** Use sample data and apply a suitable statistical test (like t-test, chi-square, ANOVA).
4. **Make a decision:**

   * If the p-value ≤ α, we reject the null hypothesis (evidence supports the alternative).
   * If the p-value > α, we fail to reject the null hypothesis (not enough evidence to support the alternative).

**Example:**
Suppose a company claims their bulb lasts 1,000 hours on average. You collect a sample of bulbs and test whether the average lifetime is really 1,000 hours. Hypothesis testing will tell you whether the data supports the company’s claim or not.

**In short:** Hypothesis testing is like a decision-making tool that uses sample data to check whether assumptions about a population are reasonable or should be rejected.





**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**
In statistics, hypotheses are statements we test using data. The two main types are:

1. **Null Hypothesis (H₀):**

   * It is the default or starting assumption.
   * It states that there is **no effect, no difference, or no relationship** in the population.
   * Example: “A new fertilizer does not increase crop yield compared to the old one.”

2. **Alternative Hypothesis (H₁ or Ha):**

   * It is the opposite of the null hypothesis.
   * It states that there **is an effect, a difference, or a relationship** in the population.
   * Example: “The new fertilizer increases crop yield compared to the old one.”

**Key Differences:**

* **Meaning:** The null assumes no change; the alternative assumes change.
* **Testing:** The null hypothesis is what we try to reject using data, while the alternative is what we try to prove.
* **Decision:**

  * If evidence (p-value) is strong, we reject H₀ and accept H₁.
  * If evidence is weak, we fail to reject H₀.

**In short:**
The null hypothesis is a “no difference” claim, while the alternative hypothesis is a “there is a difference” claim.





**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**
The **significance level**, written as **α (alpha)**, is a threshold value in hypothesis testing that tells us how much risk of error we are willing to accept when making a decision about the null hypothesis (**H₀**).

* It represents the probability of making a **Type I error** (rejecting H₀ when it is actually true).
* Common choices are **α = 0.05 (5%)**, **0.01 (1%)**, or **0.10 (10%)**.

**Role in deciding the outcome:**

1. After data is collected, a **p-value** is calculated.
2. The p-value is compared with the significance level (α):

   * If **p-value ≤ α** → Reject the null hypothesis (enough evidence for the alternative).
   * If **p-value > α** → Fail to reject the null hypothesis (not enough evidence to support the alternative).
3. The significance level controls how strict the test is:

   * **Lower α (e.g., 0.01):** Very strict, reduces false positives but harder to prove an effect.
   * **Higher α (e.g., 0.10):** Less strict, easier to prove an effect but increases the risk of error.

**Example:**
Suppose a researcher sets α = 0.05 in a drug trial. This means they are ready to accept a 5% chance of wrongly concluding that the drug is effective when it actually isn’t.

**In short:**
The significance level is the decision boundary that helps us judge whether the sample evidence is strong enough to reject the null hypothesis.





**Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer:**
In hypothesis testing, two kinds of errors can occur when making decisions about the null hypothesis (**H₀**):

### 1. **Type I Error (False Positive)**

* Occurs when we **reject H₀ even though it is actually true**.
* Probability of making a Type I error = **α (significance level)**.
* Example:
  A medical test concludes that a healthy person has a disease.

  * Here, H₀: “The person is healthy.”
  * Error: The test rejects H₀ and says “sick,” even though the person is actually healthy.

### 2. **Type II Error (False Negative)**

* Occurs when we **fail to reject H₀ even though it is actually false**.
* Probability of making a Type II error = **β (beta)**.
* Example:
  A medical test fails to detect a disease in a person who actually has it.

  * Here, H₀: “The person is healthy.”
  * Error: The test does not reject H₀, but the person is actually sick.

**Key difference:**

* **Type I Error:** Detecting something that isn’t there (false alarm).
* **Type II Error:** Missing something that is actually there (missed detection).

**In short:**
Type I error is like convicting an innocent person, while Type II error is like letting a guilty person go free.





**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**
Both **Z-test** and **T-test** are statistical methods used to test hypotheses about population means. The main difference lies in the information we have and the sample size.

### **Z-test:**

* Used when the **population variance (σ²) or standard deviation (σ)** is known.
* Works well for **large sample sizes (n ≥ 30)**.
* Assumes the data follows a **normal distribution**.
* **Example:** Testing whether the average height of students is equal to 160 cm when the population standard deviation is known.

### **T-test:**

* Used when the **population variance/standard deviation is unknown**.
* More appropriate for **small sample sizes (n < 30)**.
* Uses the **sample standard deviation (s)** as an estimate.
* Based on the **t-distribution**, which has heavier tails than the normal distribution to account for extra uncertainty.
* **Example:** Testing whether the average exam score of a small class of 20 students is equal to 70, when the population standard deviation is not known.

### **Key Differences:**

| Feature             | Z-test              | T-test                          |
| ------------------- | ------------------- | ------------------------------- |
| Population variance | Known               | Unknown (estimated from sample) |
| Sample size         | Large (≥30)         | Small (<30)                     |
| Distribution used   | Normal distribution | t-distribution                  |

**In short:**

* Use a **Z-test** when the sample size is large and the population standard deviation is known.
* Use a **T-test** when the sample size is small or the population standard deviation is unknown.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Population mean (hypothesized)
mu = 50

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # sample standard deviation
n = len(data)

# Calculate Z statistic
z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant evidence).")
else:
    print("Fail to reject the null hypothesis (no significant evidence).")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, sample size=100

# Step 2: Calculate mean and confidence interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# 95% confidence interval
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)  # z-value for 95% confidence
margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Step 3: Plot histogram of data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_zscores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.random.normal(loc=100, scale=15, size=200)  # Normal distribution

# Calculate Z-scores
z_scores = calculate_zscores(data)

# Print first 10 Z-scores
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=20, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (0)")
plt.legend()
plt.show()
```
